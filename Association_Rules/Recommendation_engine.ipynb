```python
new_transactions=new_transactions.pivot_table(index="Transaction",columns='product',values='quantity',aggfunc='sum').fillna(0)
```

```python

##Alternte methode no need

train_basket=new_transactions.groupby(['Transaction', 'product'])

"""
#its group transaction and product unique values
#ex- 1 apple
     2 orrange
     1 apple

     1 apple
       apple

     2 Orrange
"""
count=train_basket['product'].count() 

""" 
apple in transaction 1 =2
"""
count

train_basket=count.reset_index(name='count')
""" 
pivot_table:

The pivot_table method reshapes the DataFrame.
Parameters used:
index='transaction': Each row in the resulting pivot table corresponds to a unique transaction.
columns='product': Each column corresponds to a unique product.
values='Count': The values in the pivot table are taken from the Count column in the original DataFrame.
aggfunc='sum': If there are multiple rows with the same transaction and product, their Count values are summed.

"""

train_basket=train_basket.pivot_table(index='Transaction',columns='product',values='count',aggfunc='sum').fillna(0)
```

```python
new_transactions
```

```python
train_basket=new_transactions
```

```python
train_basket.head(10)
```

```python
def encode_units(x):
    if x<=0:
        return 0
    else:
        return 1

my_basket=train_basket.applymap(encode_units)
```

```python
my_basket
```

Get Frequent item sets

```python
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
frequent_items= apriori(my_basket,min_support=0.04,use_colnames=True)
```

```python
frequent_items
```

```python
"""
Lift > 1: Indicates a positive association between items 

A and 

B. They are more likely to be purchased together than by chance.
Lift = 1: Indicates that 

A and 

B are independent. Their co-occurrence is purely coincidental and not significant.
Lift < 1: Indicates a negative association. It means 

A and 

B are less likely to be bought together than expected by chance.

"""


rule_basket=association_rules(frequent_items,metric="lift",min_threshold=1,num_itemsets=len(frequent_items))
```

```python
len(rule_basket)
```

```python
# Get Example Recommendations
rule_basket[rule_basket['antecedents']=={'Cappuccino'}].sort_values('confidence',ascending=False).head(50)
```

```python
product_categories = Dataset[['product','product_category']].drop_duplicates().set_index('product').to_dict()['product_category']
```

```python
recommendations_json = {}

antecedents = rule_basket['antecedents'].unique()
for antecedent in antecedents:
    df_rec = rule_basket[rule_basket['antecedents']==antecedent]
    df_rec = df_rec.sort_values('confidence',ascending=False)
    key = "_".join(antecedent)
    recommendations_json[key] = []
    for _, row in df_rec.iterrows():
        rec_objects =row['consequents']
        for rec_object in rec_objects:
            already_exists = False
            for current_rec_object in recommendations_json[key]:
                if rec_object == current_rec_object['product']:
                    already_exists=True
            if already_exists:
                continue
            
            rec = {'product':rec_object, 
                   'product_category':product_categories[rec_object],
                   'confidence': row['confidence']
                  }
            recommendations_json[key].append(rec)
```
